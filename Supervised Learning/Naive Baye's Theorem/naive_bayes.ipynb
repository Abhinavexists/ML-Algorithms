```python
from sklearn.datasets import load_iris

x,y = load_iris(return_X_y=True)
# return_X_y=True tells load_iris to return the features (X) and 
# target labels (y) directly as separate NumPy arrays.
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train, y_train)

y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))

print("Confusion Matrix")
print(confusion_matrix(y_pred, y_test))
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(), df.info()
```

```python
# independent and dependent features
x = df[['total_bill','tip', 'sex', 'day', 'time', 'size']]
y = df['smoker']
```

```python
# Feature Encoding (Label Encoding and OneShot Encoding)
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False) # prevents dummy variable trap
categorical_cols = ['sex','day','time']
ohe_cat = ohe.fit_transform(x[categorical_cols])
```

```python
# encoded features to dataframe
import pandas as pd

ohe_df = pd.DataFrame(ohe_cat, columns=ohe.get_feature_names_out(categorical_cols))
```

```python
# Combine Numerical features
num_val = x.drop(columns=categorical_cols).reset_index(drop=True)
x_final = pd.concat([num_val, ohe_df], axis=1)
```

```python
x_final
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_final = le.fit_transform(y)
```

```python
y_final
```

```python
# Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, test_size=0.3, random_state=0)
```

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
```

```python
print(accuracy_score(y_pred, y_test))
print(classification_report(y_pred, y_test))

print("Confusion Matrix")
print(confusion_matrix(y_pred, y_test))
```
